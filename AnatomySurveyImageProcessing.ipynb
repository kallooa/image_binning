```python
img_path = '/Users/aadi/Documents/GitHub/anatomy-map/lib/img/anatomic/'
#arr = os.listdir(img_path)
included_extensions = ['jpg', 'bmp', 'png', 'gif']
arr = [fn for fn in os.listdir(img_path) if any(fn.endswith(ext) for ext in included_extensions)]

for i in range(0, len(arr)):
    print(i, arr[i])
    savename = arr[i].translate(None, digits).replace(" ", "")
    img = cv2.imread(img_path+arr[i])

    #img = cv2.blur(img,(3,3))

    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    f = image[:,:,0]
    #plt.imshow(f)

    # red color boundaries (R,B and G)
    lower = [30, 30, 30]
    upper = [255, 70, 60]

    # create NumPy arrays from the boundaries
    lower = np.array(lower, dtype="uint8")
    upper = np.array(upper, dtype="uint8")

    # find the colors within the specified boundaries and apply
    # the mask
    mask = cv2.inRange(image, lower, upper)
    gray = cv2.bitwise_and(f, f, mask=mask)
    output = image.copy()#cv2.bitwise_and(image, image, mask=mask)

    ret,thresh = cv2.threshold(gray, 126, 255, cv2.THRESH_BINARY)

    im2,contours,hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    #minX,minY,maxX,maxY = 0
    coord_array = []
    counter = 0
    if len(contours) != 0:
        for ct in contours:
            # draw in blue the contours that were founded
            cv2.drawContours(output, ct, -1, 255, 3)

            #find the biggest area
            #c = max(contours, key = cv2.contourArea)
            #print(c)

            x,y,w,h = cv2.boundingRect(ct)
            coord_array.append([cv2.boundingRect(ct)])
            # draw the book contour (in green)
            #cv2.rectangle(output,(x,y),(x+w,y+h),(0,255,0),2)
            #print(coord_array)
            #print(im2)
            #print(hierarchy)
    offset = 120
    print(len(coord_array))
    if(len(coord_array)==0):
        continue
    maxX = max([[t[0] for t in l] for l in coord_array])[0] + max([[t[2] for t in l] for l in coord_array])[0] + offset
    minX = min([[t[0] for t in l] for l in coord_array])[0] - offset
    maxY = max([[t[1] for t in l] for l in coord_array])[0] + max([[t[3] for t in l] for l in coord_array])[0] + offset
    minY = min([[t[1] for t in l] for l in coord_array])[0] - offset

    cv2.rectangle(output,(minX,minY),(maxX,maxY),(0,255,0),2)

    # show the images
    #plt.figure(figsize=(10, 10))
    #plt.title('Scalp')
    #plt.imshow(output)
    #print(contours[0])
    #plt.show()
    cropped = output[minY:maxY, minX:maxX]
    cv2.imwrite('/Users/aadi/Documents/GitHub/anatomy-map/lib/img/anatomic_processed/'+savename, cropped)
```
